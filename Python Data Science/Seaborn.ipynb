# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Scatterplot

```python
df = pd.read_csv("/Users/eduardoarmenta/Desktop/LEARN/UDEMY/Python Data Science/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/dm_office_sales.csv")
df
```

```python
# scatterplot
# since it uses matplotlib, seaborn takes in the same commands
plt.figure(dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2',size='work experience',alpha=.5).legend(loc=(1.05,0.05));
```

### Distribution plots

```python
# rugplot. height is percentage of y space taken by the lines
sns.rugplot(df['salary'], height = 1);
```

```python
# displot
# set style is a high level command
sns.set(style='darkgrid')
sns.displot(data=df['salary'],bins=25,rug=True,kde=True);
```

```python
# histogram and KDE plot
sns.histplot(data=df['salary'],kde=True,color='red',edgecolor='blue');
```

```python
# kde plot. clip decides the range to show of the kde curve
# bw adjust is for choosing how much variance you want to pick up with the curve
sns.kdeplot(data=df['salary'],clip=[0,160000],bw_adjust=0.5,shade=True);
```

### Categorical plots

```python
# countplot
plt.figure(figsize=(10,4),dpi=100)
sns.countplot(data=df,x='division',hue='level of education',palette='Set2');
# plt.ylim(0,100)
```

```python
# barplot
plt.figure(figsize=(10,4),dpi=70)
sns.barplot(data=df, x='level of education',y='salary',estimator=np.mean,ci='sd',hue='division')
plt.legend(bbox_to_anchor=(1.05,.5));
```

### Distribution within categories

```python
# load in dataframe
df1 = pd.read_csv("/Users/eduardoarmenta/Desktop/LEARN/UDEMY/Python Data Science/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/StudentsPerformance.csv")
df1
```

```python
# boxplot
sns.boxplot(data=df1,x='math score',y='parental level of education',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,0.5));
```

```python
# boxen plot
sns.boxenplot(data=df1,x='math score',y='test preparation course',hue='gender');
```

### Comparison plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("/Users/eduardoarmenta/Desktop/LEARN/UDEMY/Python Data Science/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/StudentsPerformance.csv")
df1
```

```python
# jointplots. it's a combo of scatterplot and histogram
sns.jointplot(data=df1, x='math score',y='reading score',color='purple',kind='hex');
```

```python
# jointplots. it's a combo of scatterplot and histogram
sns.jointplot(data=df1, x='math score',y='reading score',hue='gender',alpha=0.3);
```

```python
# pairplot. corner True removes the duplicate graphs
sns.pairplot(data=df1,hue='gender',corner=True);
```

### Grid plots

```python
# it's like using subplots based on a column
sns.catplot(data=df1,x='gender',y='math score',kind='box',row='gender',col='test preparation course');
```

```python
# similar to a pair grid but we decide which types of plots to use on each level
g = sns.PairGrid(df1)
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot, shade=True)
g = g.map_diag(sns.histplot)
```

### Matrix plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df2 = pd.read_csv("/Users/eduardoarmenta/Desktop/LEARN/UDEMY/Python Data Science/UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/country_table.csv")
df2
```

```python
df2 = df2.set_index('Countries')
```

```python
# heatmap. lw adds width to the lines between cells. annot adds numbers. cmap is for palette
sns.heatmap(data=df2.drop('Life expectancy', axis=1),lw=0.5,annot=True,cmap='viridis');
```

```python
# clustermap. lw adds width to the lines between cells. annot adds numbers. cmap is for palette.
# you can col_cluster=False to remove clusters by columns
sns.clustermap(data=df2.drop('Life expectancy', axis=1),lw=0.5,annot=True,cmap='viridis');
```
