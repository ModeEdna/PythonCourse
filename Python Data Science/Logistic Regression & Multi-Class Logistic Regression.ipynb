# Logistic Regression

Logistic Regression is for predicting categorical labels while Linear Regression is for continuous labels.

Theoretically, continuous labels can be turned into categories (ex: price from 0-100, 101-200, etc.).

Classification models often produce a probability prediction. In other words, how likely it is that Y is correctly predicted.

Given that it's a categorical prediction, you can't use metrics like mean squares, so we use other metrics instead.

### Classification metrics

**CONFUSION MATRIX**

Matrix with false positives, false negatives, true positives, and true negatives.

*Accuracy:* how often is the model correct? True+ + True- / all

*Recall (sensitivity):* when it actually is a positive case, how often is it correct? True+ / actual+

*Precision:* when the predicition is positive, how often is it correct? True+ / predicted+

*F-1 Score:* harmonic mean of precision and recall. (2 x precision x recall) / (precision + recall) . If F1 score is 0, either precision or recall or both are also 0.

**ROC Curves**

Shows the True vs False positives on a curve depending on the change of threshold of probability of a guess belonging to either class.

X = False positive rate

Y = True positive rate

*AUC:* Area under the curve. The more area, the better the model.

**Splitting, scaling**

```python
X = df.drop('species', axis=1)
# scikit has no problem with passing strings as classes
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
log_model = LogisticRegression(solver='saga', multi_class='ovr', max_iter=5000)
```

```python
penalty = ['l1', 'l2', 'elasticnet']
l1_ratio = np.linspace(0,1,20)
C = np.logspace(0,10,20)

param_grid = {'penalty':penalty, 'l1_ratio':l1_ratio, 'C':C}
```

```python
grid_model = GridSearchCV(log_model, param_grid=param_grid)
```

```python
grid_model.fit(scaled_X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, plot_confusion_matrix
```

```python
grid_model.best_params_
```

```python
y_pred = grid_model.predict(scaled_X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
plot_confusion_matrix(grid_model, scaled_X_test, y_test);
```

```python
print(classification_report(y_test, y_pred))
```

```python
# built-in plot ROC curve doesn't work with multi-class models
from sklearn.metrics import roc_curve, auc
```

```python
def plot_multiclass_roc(clf, X_test, y_test, n_classes, figsize=(5,5)):
    y_score = clf.decision_function(X_test)

    # structures
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    # calculate dummies once
    y_test_dummies = pd.get_dummies(y_test, drop_first=False).values
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_dummies[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # roc for each class
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('Receiver operating characteristic example')
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f) for label %i' % (roc_auc[i], i))
    ax.legend(loc="best")
    ax.grid(alpha=.4)
    sns.despine()
    plt.show()
```

```python
plot_multiclass_roc(grid_model, scaled_X_test, y_test, n_classes=3)
```
